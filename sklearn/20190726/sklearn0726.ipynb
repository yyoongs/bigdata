```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
import pandas as pd
```

```python
iris = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
data
```

```python
iris_target = pd.DataFrame(data.target)
```

```python
iris_target = pd.DataFrame(data.target, columns=['target'])
```

```python
iris = pd.concat([iris,iris_target],axis=1)
```

```python
iris
```

```python
from sklearn.model_selection import train_test_split
# 트레이닝 데이터와 테스트 데이터를 쪼개줌!
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(iris.iloc[:,:-1],iris.iloc[:,-1])
```
