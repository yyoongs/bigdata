fbprophet

시계열 분석 알고리즘 / 쓰기 쉬움!

설치에 종속성이 몇개 있어서 설치가 어려움

```python
from fbprophet import Prophet
```

```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/master/examples/example_wp_log_peyton_manning.csv')
```

```python
data
```

```python
m = Prophet()
```

```python
m.fit(data)
```

```python
vars(m)
```

```python
future = m.make_future_dataframe(365) # 미래 데이터 만들기
```

```python
vars(future)
```

```python
future
```

```python
forecast = m.predict(future) # 미래 데이터 예측 하기
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
m.plot(forecast)
```

```python
m.plot_components(forecast)
```
