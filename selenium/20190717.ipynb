```python
import requests

def download(method, url, params=None, data=None, headers=None):
    try:
        resp = requests.request(method, url, params=params,data=data,headers=headers)
    except requests.exceptionsp.HTTPerror as e:
        print(e.code)
        print(e.reason)
    
    return resp
```

```python
def parseURL(seed):
    html = download("get", seed)
    dom = BeautifulSoup(html.text, "lxml")
    
    return [requests.compat.urljoin(url, _["href"]) for _ in dom.find_all("a") if _.has_attr("href") and len(_["href"]) > 3]
```

```python
con = sqlite3.connect("bot3.db")
cur = con.cursor()

cur.executescript("""
    DROP TABLE IF EXISTS table1;
    CREATE TABLE table1(
        id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
        title TEXT NOT NULL,
        url TEXT NOT NULL,
        date TIMESTAMP DEFAULT CURRENT_TIMESTAMP NOT NULL
    );

    DROP TABLE IF EXISTS table2;
    CREATE TABLE table2(
        id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
        table1_id INTEGER NOT NULL,
        content TEXT NOT NULL,
        date TIMESTAMP DEFAULT CURRENT_TIMESTAMP NOT NULL
    );
   
""")
```

```python
driver.get("https://news.naver.com/")
container2 = driver.find_elements_by_css_selector("div.hdline_article_tit > a , .com_list li > a")

for _ in container2:
    print(_.get_attribute("href"))
    
urlList = [_.get_attribute("href") for _ in container2 if len(_.get_attribute("href")) > 3]
print(urlList)

for url in urlList:
    driver.execute_script("javascript:window.open('about:blank')")
    driver.switch_to_window(driver.window_handles[-1])
    driver.get(url)
    title = driver.find_element_by_id("articleTitle")
    content = driver.find_element_by_id("articleBodyContents")
    
    print(type(title),type(content))
    cur.execute("""
        INSERT INTO table3 (title, content, url)
        VALUES (?,?,?)
    """, [title.text, content.text, url])
    con.commit()
    time.sleep(1)
    driver.close()
```
